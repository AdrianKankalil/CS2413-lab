## Using Webscraping to Extract Data


Use the `requests` library to download the webpage. Save the text of the response as a variable named `html_data`.


```python
# Your code here:

url = "https://www.example.com"
html_data = requests.get(url).text
```

Parse the html data using `beautiful_soup`.


```python
soup = BeautifulSoup(html_data, "html5lib")
```

<b>Question 1</b> what is the content of the title attribute:


```python
# Your code here:

page_title = soup.title.string
page_title
```

<b>Question 2</b> Extract the link in this webpage using "find_all" method.


```python
# Your code here:
links = soup.find_all("a")

for link in links:
    print(link.get("href"))

```
